- color --> red, blue , green etc, hex codes for colors (#FF0000 --- red)

- line width (lw): pass any numeric value
- marker : 'o' (circle), 's' (square), '^' (triangle), '*' (stars), '.' (point) etc
- markerfacecolor : red, blue , green etc, hex codes for colors (#FF0000 --- red)
- markersize : any numerical value

- ameker edge color : red, blue , green etc, hex codes for colors (#FF0000 --- red)

#### Object Oriented Approach

- This approach is also known as figure or axes approach
- In this approach, we can increase and decrease the size of the whole figure or a particular axes

#### How to create a figure object



```python
plt.figure()
```

#### figure_object.add_axes([left_boundary, bottom_boundary, width, height])

```python
f = plt.figure()
f.add_axes([.1, .1, .5, .5])
f.add_axes([.7, .1, .6, .8])
```

#### How to plot the data in a particular axes



```python
f = plt.figure()
a = f.add_axes([.1,.1,.5,.5])
b = f.add_axes([.7,.1,.5,.5])
c = f.add_axes([.7,.7,.5,.5])
d = f.add_axes([.1,.7,.5,.5])

a.plot(x,y, color = 'r')
b.plot(y, x)
c.plot(x**2, y)
d.plot([10, 15, 9], [7,7,7], color = 'r', marker = 'o')

plt.show()
```

### How to give title and axes name in the object oriented approach

    Syntax:

            set_title('title_name')
            set_xlabel('X_axis_name')
            set_xlabel('Y_axis_name')

```python
arr = np.arange(10)
arr
```

```python
arr2 = arr*2
arr2
```

```python
f = plt.figure()
a = f.add_axes([.1,.1,.5,.5])
b = f.add_axes([.7,.1,.5,.5])

a.plot(arr, arr2, color = 'r', marker = 'o')
a.set_title('Plot-1')
a.set_xlabel('arr')
a.set_ylabel('arr2')

b.plot(arr, arr**2, color = 'y', marker = 'o')
b.set_title('Plot-2')
b.set_xlabel('arr')
b.set_ylabel('arr3')
```

### How to increase and decrease the size of the figure

    Syntax:

            plt.figure(figsize = (width, height), dpi = int)

            where dpi is dots per inches

```python
f = plt.figure(figsize = (5,5), dpi = 80)
a = f.add_axes([.1,.1,.5,.5])
b = f.add_axes([.7,.1,.5,.5])

a.plot(arr, arr2, color = 'r', marker = 'o')
a.set_title('Plot-1')
a.set_xlabel('arr')
a.set_ylabel('arr2')

b.plot(arr, arr**2, color = 'y', marker = 'o')
b.set_title('Plot-2')
b.set_xlabel('arr')
b.set_ylabel('arr3')
```
