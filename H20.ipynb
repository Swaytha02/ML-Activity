```python
import h2o
```

```python
h2o.init()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train_h2o = h2o.H2OFrame(train)
test_h2o = h2o.H2OFrame(test)
```

```python
x = train_h2o.columns[:-1]
y = train_h2o.columns[-1]
```

```python
from h2o.automl import H2OAutoML
aml = H2OAutoML(max_models=10, seed=1)
```

```python
aml.train(x=x, y=y, training_frame=train_h2o)
```

```python
leaderboard = aml.leaderboard
print(leaderboard)
```

```python
# Print all rows instead of 10 rows
leaderboard.head(rows = leaderboard.nrows)
```

```python
best_model = aml.leader
```

```python
predictions = best_model.predict(test_h2o)
```

```python
df = pd.read_csv('/content/data.csv')
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
rg.fit(x_train,y_train)
```

```python
import gradio as gr
```

```python
import numpy as np
```

```python
def greet(number_courses,time_study):
    # Assuming rg is a trained linear regression model
    linear_prediction = rg.predict(np.array([float(number_courses),float(time_study)]).reshape(1,-1))
    return linear_prediction[0]
```

```python
# Create a Gradio interface
demo = gr.Interface(fn=greet,
                    inputs=[gr.Textbox(label="number_courses"),
                            gr.Textbox(label="time_study")],
                    outputs=gr.Textbox(label="Marks"),
                    title="Student marks evaluation",
                    description="Evaluation of students Performance using past data"
                   )

# Launch the interface
demo.launch(debug=True)
```
